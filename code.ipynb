In this implementation, we solve the differential equation
dy/dt = cos(t) - y, with initial condition y(0) = 1,
on the interval [0, 5] using Euler’s method with n = 20 steps.

We plot:

- The exact solution in green,

- The Euler’s method approximation in blue dashed lines with markers,

- The absolute error (difference between the exact and Euler solution) in red dashed-dot line.

The graph visually compares the numerical and exact solutions and illustrates the error across the interval.

```python
import numpy as np
import matplotlib.pyplot as plt
a, b = 0, 5
n = 20
h = (b - a) / n
t = np.linspace(a, b, n+1)
y_euler = np.zeros(n+1)
y_euler[0] = 1
def f(t, y):
    return np.cos(t) - y
for i in range(n):
    y_euler[i+1] = y_euler[i] + h * f(t[i], y_euler[i])
def y_exact(t):
    return 0.5*(np.cos(t) + np.sin(t)) + 0.5*np.exp(-t)
y_orig = y_exact(t)
error = np.abs(y_orig - y_euler)
plt.figure(figsize=(10,6))
plt.plot(t, y_orig, 'g-', label='Exact Solution')
plt.plot(t, y_euler, 'b--o', label="Euler's Method")
plt.plot(t, error, 'r-.', label='Absolute Error')
plt.title("Euler's Method vs Exact Solution")
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
